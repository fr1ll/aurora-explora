```python
import pandas as pd
from PIL import Image
from pathlib import Path
from clip_interrogator import Interrogator, Config
```

```python
base_path = Path("clean_aurs_04")
df = pd.read_parquet(base_path/"triangles/imgpaths-with-triangles.parquet")
```

```python
im = Image.open(df["image_path"].values[0])
```

```python
ci = Interrogator(Config(clip_model_name="ViT-H-14/laion2b_s32b_b79k"))
```

```python
im_cap = ci.interrogate(im)
print(im_cap)
```

```python
captions=[]
for p in list(df["image_path"].values):
    i = Image.open(p)
    captions += [ci.interrogate(i)]
```

```python
df["captions"] = captions
df.write_parquet(base_path/"triangles/imgpaths-triangles-captions.parquet")
```
